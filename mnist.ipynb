```python
import matplotlib.pyplot as plt
from skimage import transform
import numpy as np
```

**MNIST** **DATASET**

 MNIST dataset is one of the most famous datasets in the field of machine learning and computer vision. It stands for Modified National Institute of Standards and Technology database. MNIST consists of 70,000 grayscale images of handwritten digits (0 through 9), with each image being a 28x28 pixel square.

```python
from keras.datasets import mnist,fashion_mnist,cifar10,cifar100
```

```python
#loading the dataset and spliting the data into training and testing data
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
#importing the model
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
imrecog1=RandomForestClassifier(n_estimators=10)
```

```python
#preprocessing the input data
```

```python
X_test=[]
for x in x_test:
  rescale=transform.resize(x,(100,100))
  X_test.append(rescale)
X_test=np.array(X_test)
```

```python
#training the model with training data
```

```python
imrecog1.fit(x_train.reshape(len(x_train),-1),y_train)
```

```python
  # testing the model with test data
```

```python
y_pred=imrecog1.predict(x_test.reshape(len(x_test),-1))
```

```python
# model evaluation
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
# data visualization
```

```python
plt.figure(figsize=(10,5))
for i in range(8):
  plt.subplot(2,4,i+1)
  plt.imshow(X_test[i],cmap="gray")
  plt.axis("off")
  plt.title(f"Predicted:{y_pred[i]}\nActual:{y_test[i]}")
plt.tight_layout()
plt.show()
```
