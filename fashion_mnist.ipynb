```python
import matplotlib.pyplot as plt
from skimage import transform
import numpy as np
```

**Fashion MNIST dataset**

Fashion MNIST dataset consists of images of clothing items instead of handwritten digits. It was created as a more challenging alternative to MNIST for benchmarking machine learning algorithms, particularly in the realm of image classification.
Fashion MNIST contains 70,000 grayscale images, also divided into 60,000 training examples and 10,000 testing examples. Each image is a 28x28 pixel square, similar to MNIST. However, instead of digits, the images represent 10 different classes of fashion items, including:

1.T-shirt/top

2.Trouser

3.Pullover

4.Dress

5.Coat

6.Sandal

7.Shirt

8.Sneaker

9.Bag

10.Ankle boot

```python
from keras.datasets import fashion_mnist
```

```python
# loading the dataset and spliting the data into training and testing data
```

```python
(x_train,y_train),(x_test,y_test)=fashion_mnist.load_data()
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
X_test=[]
for x in x_test:
  rescale=transform.resize(x,(100,100))
  X_test.append(rescale)
X_test=np.array(X_test)
```

```python
#importing the model
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
imrecog2=RandomForestClassifier(n_estimators=10)
```

```python
# traininig the model with training data
```

```python
imrecog2.fit(x_train.reshape(len(x_train),-1),y_train)
```

```python
# testing the model with test data
```

```python
y_pred=imrecog2.predict(x_test.reshape(len(x_test),-1))
```

```python
# model evaluation
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
# data visualization
```

```python
label_names=[ 'T-shirt/top', 'Trouser','Pullover','Dress','Coat','Sandal', 'Shirt','Sneaker', 'Bag','Ankle boot']
```

```python
plt.figure(figsize=(10,5))
for i in range(8):
  plt.subplot(2,5,i+1)
  plt.imshow(X_test[i],cmap="gray")
  plt.axis("off")
  plt.title(f"Predicted:{label_names[y_pred[i]]}\nActual:{label_names[y_test[i]]}")
plt.tight_layout()
plt.show()
```
